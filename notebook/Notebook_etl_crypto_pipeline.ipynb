# Crypto ETL Pipeline (Demo)


This notebook demonstrates an end-to-end ETL pipeline that extracts cryptocurrency market data, transforms it into a structured format, and loads it into a CSV file.

Built as part of a Data Engineering portfolio project.


##  Project Overview

This ETL pipeline performs the following steps:

- **Extract**: Fetch cryptocurrency data from the CoinGecko API
- **Transform**: Clean and structure the data using pandas
- **Load**: Save the processed data into a CSV file

This project simulates a real-world data engineering workflow.

## Tech Stack

- Python
- Pandas
- Requests
- CSV (data storage)
- Google Colab (execution environment)

## Step 1: Extract Data

In this step, we fetch cryptocurrency market data from the CoinGecko API.

```python
%%writefile extract.py

import requests

def extract_data():
    url = "https://api.coingecko.com/api/v3/coins/markets"

    params = {
        "vs_currency": "usd",
        "order": "market_cap_desc",
        "per_page": 10,
        "page": 1
    }

    response = requests.get(url, params=params)

    if response.status_code == 200:
        data = response.json()
        print("Data extracted successfully!")
        return data
    else:
        print("Error fetching data")
        return None
```

## Step 2: Transform Data

We convert the raw JSON data into a structured pandas DataFrame and select relevant columns.

```python
%%writefile transform.py

import pandas as pd

def transform_data(data):
    if not data:
        print("No data to transform")
        return None

    df = pd.DataFrame(data)

    # seleccionar columnas importantes
    df = df[[
        "id",
        "symbol",
        "name",
        "current_price",
        "market_cap",
        "total_volume"
    ]]

    print("Data transformed successfully!")
    return df
```

## Step 3: Load Data

The transformed data is saved into a CSV file for further analysis or downstream use.

```python
%%writefile load.py

import os

def load_data(df):
    if df is None:
        print("No data to load")
        return

    output_path = "data/crypto_data.csv"

    # crear carpeta si no existe
    os.makedirs("data", exist_ok=True)

    df.to_csv(output_path, index=False)

    print(f"Data saved to {output_path}")
```

## Run ETL Pipeline

Now we execute the full pipeline end-to-end.

```python
%%writefile main.py

from extract import extract_data
from transform import transform_data
from load import load_data


def main():
    print("Starting ETL pipeline...")

    data = extract_data()

    if data:
        print(f"Extracted {len(data)} records")

        df = transform_data(data)

        if df is not None:
            load_data(df)


if __name__ == "__main__":
    main()
```
